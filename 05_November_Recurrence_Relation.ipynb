1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.

T(2)
= 3T(1) + 12 * 2
= 3(3T(0) + 12 * 1) + 12 * 2
= 3^2 T(0) + 3 * 12 * 1 + 12 * 2
= 3^2 * 5 + 36 + 24
= 9 * 5 + 60
= 45 + 60
= 105
​
T(2)=105.


To find T(1), we substitute n=1 into the recurrence relation:
= 3T(0) + 12 * 1
= 3 * 5 + 12
= 15 + 12
= 27
​
 
So, the correct value is 
T(1)=27.
